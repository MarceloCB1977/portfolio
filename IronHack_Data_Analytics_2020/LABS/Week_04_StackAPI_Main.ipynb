#### Question 1: Find the questions and answers of last month.

```python
# your code here
from datetime import datetime
SITE = StackAPI('stackoverflow')
```

```python
questions = SITE.fetch('questions', fromdate=datetime(2020,2,1), todate=datetime(2020,2,29))
```

```python
questions
```

```python
answer = SITE.fetch('answers', fromdate=datetime(2020,2,1), todate=datetime(2020,2,29))
```

```python
answer
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions2 = SITE.fetch('questions', fromdate=datetime(2020,2,1), todate=datetime(2020,2,29),
    min=5, tagged='python', sort='votes')
```
