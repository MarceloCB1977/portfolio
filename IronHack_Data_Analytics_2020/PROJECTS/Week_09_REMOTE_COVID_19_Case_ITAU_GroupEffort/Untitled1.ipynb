```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cities_in_Australia_by_population'
aust = BeautifulSoup(requests.get(url).content)
```

```python
tables = pd.read_html(url)
tables
```

```python
aust.find_all('th')[:7]
head = [x.text.replace('\n', '').strip() for x in aust.find_all('th')[:7]]
head
```

```python
aust.find_all('td', attrs={})
```

```python
aust_title = aust.find_all('td')
aust_title
```

```python
cities_list = []

for each_row in aust.find_all('td'):
    links = each_row.find('a', href=True)
    if links:  # check before you access
        title = links.get('title')
        cities_list.append(title)
cities_list
```
