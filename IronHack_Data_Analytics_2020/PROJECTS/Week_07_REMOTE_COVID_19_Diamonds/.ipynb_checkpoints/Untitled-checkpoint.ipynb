```python
def cat_cut(x):
    """
    Function to categorize the types of diamond cut
    
    """
    if ((x.upper() == 'IDEAL') | ( x.upper() =='PREMIUM')):
        return '1'
    elif ((x.upper() == 'VERY GOOD') | ( x.upper() =='GOOD')):
        return '2'
    elif ((x.upper() == 'FAIR')):
        return '3'
```

```python
def cat_color(x):
    """
    Function to categorize the types of diamond color
    
    """
    if ((x.upper() == 'D') | ( x.upper() =='E') | ( x.upper() =='F')):
        return '1'
    elif ((x.upper() == 'G') | ( x.upper() =='H') | ( x.upper() =='I') | ( x.upper() =='J')):
        return '2'
    elif ((x.upper() == 'K') | (x.upper() == 'L') | ( x.upper() =='M')):
        return '3'
    elif ((x.upper() == 'N') | (x.upper() == 'O') | (x.upper() == 'P') | ( x.upper() =='Q') | ( x.upper() =='R')):
        return '4'
    elif ((x.upper() == 'S') | (x.upper() == 'T') | (x.upper() == 'U') | ( x.upper() =='V') | ( x.upper() =='X') | ( x.upper() =='Y') | ( x.upper() =='Z')):
        return '5'
```

```python
def cat_clarity(x):
    """
    Function to categorize the types of diamond clarity
    
    """
    if x.upper() == 'IF':
        return '1'
    elif ((x.upper() == 'VVS1') | ( x.upper() =='VVS2')):
        return '2'
    elif ((x.upper() == 'VS1') | (x.upper() == 'VS2')):
        return '3'
    elif ((x.upper() == 'SI1') | (x.upper() == 'SI2') | (x.upper() == 'SI3')):
        return '4'
    elif ((x.upper() == 'I1') | (x.upper() == 'I2') | (x.upper() == 'I3')):
        return '5'
    
```
